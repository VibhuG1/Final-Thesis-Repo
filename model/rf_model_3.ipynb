```python
def calculate_r_squared_from_correlation(y_true, y_pred):
    # Calculate the correlation coefficient
    correlation = np.corrcoef(y_true, y_pred)[0, 1]
    # Calculate R-squared using the correlation
    r_squared = correlation ** 2
    return r_squared
    
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    
def evaluate_model(model, X_train, X_test, Y_train, Y_test, transformation):
    # Predict on the training and testing sets
    Y_train_pred = model.predict(X_train)
    Y_test_pred = model.predict(X_test)
    if transformation == 'log1p':
        Y_train_pred= np.expm1(Y_train_pred)
        Y_test_pred= np.expm1(Y_test_pred)
        Y_train= np.expm1(Y_train)
        Y_test= np.expm1(Y_test)

    # Calculate RMSE, MAE, R^2, and MAPE for training set
    rmse_train = np.sqrt(mean_squared_error(Y_train, Y_train_pred))
    mae_train = mean_absolute_error(Y_train, Y_train_pred)
    r2_train = calculate_r_squared_from_correlation(Y_train, Y_train_pred)
    mape_train = mean_absolute_percentage_error(Y_train, Y_train_pred)

    # Calculate RMSE, MAE, R^2, and MAPE for testing set
    rmse_test = np.sqrt(mean_squared_error(Y_test, Y_test_pred))
    mae_test = mean_absolute_error(Y_test, Y_test_pred)
    r2_test = calculate_r_squared_from_correlation(Y_test, Y_test_pred)
    mape_test = mean_absolute_percentage_error(Y_test, Y_test_pred)

    # Create a DataFrame with the results
    # results = pd.DataFrame({
    #     "Train": [rmse_train, mae_train, r2_train, mape_train],
    #     "Test": [rmse_test, mae_test, r2_test, mape_test]
    # }, index=["RMSE", "MAE", "R2", "MAPE"])

    results = pd.DataFrame({
        "Train": [rmse_train, mae_train, mape_train,r2_train],
        "Test": [rmse_test, mae_test, mape_test,r2_test]
    }, index=["RMSE", "MAE", "MAPE", "R2"])

    return results

def get_aggregated_importances_df(model_stat_rf,cat_columns):
    aggregated_importances= {}
    # Aggregate the importances for each original categorical column
    for cat_column in cat_columns:
        # Filter the rows in 'model_stat' that belong to the current categorical column
        cat_rows = model_stat_rf[model_stat_rf['Model Parameters'].str.startswith(cat_column)]
        # Calculate the sum of importances for this categorical column
        cat_importance_sum = cat_rows['Weightage'].sum()

        # Store the aggregated importance in the dictionary
        aggregated_importances[cat_column] = cat_importance_sum
        model_stat_rf= model_stat_rf[~(model_stat_rf['Model Parameters'].str.contains(cat_column))]
    # Convert the dictionary to a DataFrame for easier analysis or visualization
    aggregated_importances_df = pd.DataFrame(list(aggregated_importances.items()), columns=['Model Parameters', 'Weightage'])
    model_stat_rf= pd.concat([model_stat_rf,aggregated_importances_df])
    model_stat_rf.sort_values('Weightage',ascending=False,inplace=True)
    return model_stat_rf
```

```python
# Create the RandomForestRegressor
rf_regressor = RandomForestRegressor(random_state=42)

# Define the parameter grid to search over
param_grid = {
    'n_estimators': [100, 200, 250],
    'max_depth': [7,8, 9],
    'min_samples_leaf': [ 4,5, 7],
    'max_features': [0.6, 0.65, 0.7],
    'bootstrap': [True, False],
}

# Define a custom scoring function for MAE
# scoring = make_scorer(lambda y_true, y_pred: -mean_absolute_error(y_true, y_pred))
scoring = make_scorer(lambda y_true, y_pred: mean_absolute_percentage_error(y_true, y_pred))

# Create the GridSearchCV object
grid_search = GridSearchCV(estimator=rf_regressor, param_grid=param_grid, cv=None, n_jobs=-1, scoring=scoring) #'neg_mean_squared_error'

# Fit the grid search to your training data
grid_search.fit(X_train, y_train)

# Get the best parameters and the best estimator
best_params = grid_search.best_params_
best_estimator = grid_search.best_estimator_

print("Best Parameters:", best_params)
print("Best Estimator:", best_estimator)
```
